```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from time import*
browser = webdriver.Chrome()
url = "http://rate.am/en/armenian-dram-exchange-rates/banks/non-cash"
browser.get(url)
all_data=browser.find_element_by_xpath('//*[@id="aspnetForm"]/div[3]/div[2]/div[1]/div[4]/div[3]/table/tbody/tr/td[1]/a')
all_data.click()
```

```python
jul_20=browser.find_element_by_xpath('//*[@id="calBorder"]/span/table/tbody/tr[3]/td/table/tbody/tr[5]/td[4]')
jul_20.click()
```

```python
show=browser.find_element_by_xpath('//*[@id="ctl00_Content_RB_dtpick1_btnOk"]')
show.click()
```

```python
import pandas as pd
import csv
data = pd.read_html(browser.find_element_by_class_name("rb").get_attribute('outerHTML'))
data = data[0]
data = data[2:19]
del data[0]
del data[2]
del data[3]
data.columns = ["name", "date", "USD_buy", "USD_sell", "EUR_buy", "EUR_sell", "RUB_buy", "RUB_sell", "GBP_buy", "GBP_sell"]
from unicodecsv import*
with open('rates_1.csv', 'w') as f:
            data.to_csv(f, header=True, index=False, encoding='utf-8')
```
